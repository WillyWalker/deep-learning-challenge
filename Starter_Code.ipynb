## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ["EIN", "NAME"])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
print( application_df["APPLICATION_TYPE"].value_counts() )
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = []
num = application_df["APPLICATION_TYPE"].unique()
cutoffVal = 500

for i in range(0, len(num)):
  if application_df["APPLICATION_TYPE"].value_counts()[num[i]] <= cutoffVal:
    application_types_to_replace.append(num[i])


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
app_df = application_df

##bins = [0, 1, 20000]
##labels = [0, 'low', '']

app_df.groupby("CLASSIFICATION").filter(lambda x: len(x) > 1)["CLASSIFICATION"].value_counts()


##app_df.head()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
num = application_df["CLASSIFICATION"].unique()
cutoffVal = 1000

for i in range(0, len(num)):
  if application_df["CLASSIFICATION"].value_counts()[num[i]] <= cutoffVal:
    classifications_to_replace.append(num[i])

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

application_df["SPECIAL_CONSIDERATIONS"] = pd.get_dummies(application_df["SPECIAL_CONSIDERATIONS"], drop_first=True, dtype='int')

USE_CASE = pd.get_dummies(application_df["USE_CASE"], dtype='int')
CLASSIFICATION = pd.get_dummies(application_df["CLASSIFICATION"], dtype='int')
APPLICATION_TYPE = pd.get_dummies(application_df["APPLICATION_TYPE"], dtype='int')
ORGANIZATION = pd.get_dummies(application_df["ORGANIZATION"], dtype='int')
AFFILIATION = pd.get_dummies(application_df["AFFILIATION"], dtype='int')
INCOME_AMT = pd.get_dummies(application_df["INCOME_AMT"], dtype='int')


application_df = application_df.merge(USE_CASE,left_index=True, right_index=True)
application_df = application_df.merge(CLASSIFICATION,left_index=True, right_index=True)
application_df = application_df.merge(APPLICATION_TYPE,left_index=True, right_index=True)
application_df = application_df.merge(ORGANIZATION,left_index=True, right_index=True)
application_df = application_df.merge(AFFILIATION,left_index=True, right_index=True)
application_df = application_df.merge(INCOME_AMT,left_index=True, right_index=True)


application_df = application_df.drop(columns=["USE_CASE", "CLASSIFICATION", "APPLICATION_TYPE", "ORGANIZATION", "AFFILIATION", "INCOME_AMT"])

application_df.head()
```

```python
import matplotlib.pyplot as plt

# Split our preprocessed data into our features and target arrays
y = application_df.IS_SUCCESSFUL.values
X = application_df.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  64
hidden_nodes_layer2 = 32
hidden_nodes_layer3 = 16
hidden_nodes_layer4 = 8
hidden_nodes_layer5 = 4
hidden_nodes_layer6 = 2


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="relu", input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Fourth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))

# Fifth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer5, activation="relu"))

# Sixth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer6, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

```

```python
# Export our model to HDF5 file

fit_model.save_weights('my_model.hdf5')
```
